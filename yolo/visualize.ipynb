```python
import numpy as np
import matplotlib.pyplot as plt
import os.path
```

```python
metrics_labels = ["Precision", "Recall", "mAP@0.5", "mAP@0.5:0.95"]
indices = [i for i in range(1, 101)]
models = ["yolov7", "yolov8", "yolov9"]
```

```python
def load_metrics(model: str):
    if os.path.isfile(f"{model}.txt"):
        return np.loadtxt(f"{model}.txt", usecols=[8, 9, 10, 11], ndmin=2).T
    else:
        return np.genfromtxt(f"{model}.csv", delimiter=",", usecols=[5, 6, 7, 8], ndmin=2, skip_header=True).T
```

```python
metrics_list = np.array([load_metrics(model) for model in models])
```

```python
def visualize(i):
    label = metrics_labels[i]
    plt.title(label)
    plt.xlabel("Epochs")
    plt.ylabel(label)
    for metrics in metrics_list:
        plt.plot(indices, metrics[i], marker='.', label=label)
    plt.legend(models, loc='lower right')
```

```python
visualize(0)
```

```python
visualize(1)
```

```python
visualize(2)
```

```python
visualize(3)
```
